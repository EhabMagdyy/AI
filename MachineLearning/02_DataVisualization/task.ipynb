```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('flights')
df.info()
df.describe()
df.head()
```

```python
sns.lineplot(x='year', y='passengers', data=df, color='purple', linewidth=3)
plt.show()
```

```python
df2 = sns.load_dataset('tips')
sns.scatterplot(data=df2, x='total_bill', y='tip', color='black')
```

```python
sns.barplot(data=df2, x='day', y='tip', estimator=sum, hue='time')  # estimator by deafult avg, hue to distingush them by a thing
plt.show()
```

```python
plt.hist(df2['total_bill'], color='lightblue', edgecolor='black', bins=30)
plt.show()
```

```python
sns.histplot(df2['total_bill'], bins=20, kde=True)
plt.show()
```

```python
sns.boxplot(data=df2, x='day', y='total_bill', color='pink')
plt.show()
```

```python
sns.countplot(x='day', data=df2, color='grey')
plt.show()
```
