```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, root_mean_squared_error,mean_squared_error,r2_score
```

```python
np.random.seed(42)
X=np.random.rand(100,1)*10
X
```

```python
y_act=2.5*X**2+0.5*X+np.random.randn(100,1)
```

```python
y_act
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y_act,test_size=0.2,random_state=42)
```

```python
# Transform features from degree 1 to degree 20
poly_features=PolynomialFeatures(degree=20)
X_train_poly=poly_features.fit_transform(X_train)
X_test_poly=poly_features.transform(X_test)
X_train_poly.shape
```

```python
model=LinearRegression()
model.fit(X_train_poly,y_train)
```

```python
y_train_pred=model.predict(X_train_poly)    # For testing overfitting (predict train data)
y_test_pred=model.predict(X_test_poly)      # For testing generalization (predict test data)
```

```python
print(f"Train Root Mean Squared Error: {root_mean_squared_error(y_train,y_train_pred)}")
print(f"Test Root Mean Squared Error: {root_mean_squared_error(y_test,y_test_pred)}")
```

```python
print(f"Train Absolute Error: {mean_absolute_error(y_train,y_train_pred)}")
print(f"Test Absolute Error: {mean_absolute_error(y_test,y_test_pred)}")
```

```python
print(f"Train MSE: {mean_squared_error(y_train,y_train_pred)}")
print(f"Test MSE: {mean_squared_error(y_test,y_test_pred)}")
```

```python
lasso_model = Lasso(alpha=0.1)  # alpha is the regularization strength
lasso_model.fit(X_train_poly, y_train)

y_train_pred_lasso = lasso_model.predict(X_train_poly)
y_test_pred_lasso = lasso_model.predict(X_test_poly)
```

```python
print(f"Train MSE: {mean_squared_error(y_train,y_train_pred_lasso)}")
print(f"Test MSE: {mean_squared_error(y_test,y_test_pred_lasso)}")
```

```python
ridge_model = Ridge(alpha=0.1)  # alpha is the regularization strength
ridge_model.fit(X_train_poly, y_train)

y_train_pred_ridge = ridge_model.predict(X_train_poly)
y_test_pred_ridge = ridge_model.predict(X_test_poly)
```

```python
print(f"Train MSE: {mean_squared_error(y_train,y_train_pred_ridge)}")
print(f"Test MSE: {mean_squared_error(y_test,y_test_pred_ridge)}")
```
