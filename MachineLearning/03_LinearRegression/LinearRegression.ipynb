```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression      
from sklearn.metrics import mean_absolute_error      
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
# Label Encoding
binary_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
df[binary_columns] = df[binary_columns].replace({'yes': 1, 'no': 0})
```

```python
df.head()
```

```python
# one hot encoding
df = pd.get_dummies(df, columns=['furnishingstatus'])
```

```python
df.head()
```

```python
# Splitting the data into features and target
y = df['price']     # Price is the target variable
x = df.drop('price', axis=1)    # All other columns are features - Separate target of the features
```

```python
x
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2, random_state=42)   # 20% Test and the other 80% Train, shuffle test set each time
```

```python
scaler = StandardScaler()
numerical_cols = ['area', 'bedrooms', 'bathrooms', 'stories']
```

```python
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols]) # train and then transform (Scale)
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])       # only transform
```

```python
X_train
```

```python
X_test
```

```python
model = LinearRegression()
```

```python

model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f"Mean Absolute Error: {mean_absolute_error(Y_test, y_pred):.2f}")
```
