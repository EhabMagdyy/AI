```python
import yaml
import os

dataset_path = os.path.abspath("ITI-Ayman-Ehab-2")

# Build corrected yaml
data = {
    'path': dataset_path,
    'train': os.path.join(dataset_path, 'train', 'images'),
    'val':   os.path.join(dataset_path, 'valid', 'images'),
    'test':  os.path.join(dataset_path, 'test', 'images'),
    'nc': 5,
    'names': ['Fist', 'Open_Palm', 'Peace', 'Thumbs_Down', 'Thumbs_Up']
}

with open(f"{dataset_path}/data.yaml", "w") as f:
    yaml.dump(data, f, default_flow_style=False)

print("Fixed data.yaml:")
with open(f"{dataset_path}/data.yaml", "r") as f:
    print(f.read())
```

```python
model = YOLO("yolov8n.pt")

results = model.train(
    data=f"{dataset_path}/data.yaml",
    epochs=20,
    imgsz=640,
    batch=8,
    name="iti_gesture_model",
    patience=10
)

print(f"Best model: {results.save_dir}/weights/best.pt")
```
