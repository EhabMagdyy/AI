```python
from ultralytics import YOLO
```

```python
DATA_YAML = "coco128.yaml"  # what does coco predict? 80 classes 
MODEL_BASE = "yolov8n.pt"
model = YOLO(MODEL_BASE)
```

```python
model.train(data=DATA_YAML, epochs=10, imgsz=640, batch=16, patience=5, project="runs/detect", name="yolov8n_coco128", exist_ok=True)
```

```python
results = model.predict(
    source="https://ultralytics.com/images/bus.jpg",
    imgsz=640,
    conf=0.25,
    save=True
)
results[0].show()
```
