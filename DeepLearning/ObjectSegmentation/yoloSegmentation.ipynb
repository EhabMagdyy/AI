```python
from ultralytics import YOLO
from pathlib import Path
```

```python
MODEL_NAME = Path("yolo11n-seg.pt")
DATASET_YAML = "coco128-seg.yaml"
test_source = "https://ultralytics.com/images/bus.jpg"
PROJECT_DIR = "runs/segment"
model = YOLO(MODEL_NAME)
```

```python
results = model.predict(
    source=test_source,
    imgsz=640,
    conf=0.5,
    save=True,
    project=PROJECT_DIR,
    name="pretrained-segmentation",
)
```

```python
model.train(
    data=DATASET_YAML,
    epochs=20,
    imgsz=640,
    batch=8,
    project=PROJECT_DIR,
    name="yolo11n-segmentation",
    pretrained=MODEL_NAME,
)
```
